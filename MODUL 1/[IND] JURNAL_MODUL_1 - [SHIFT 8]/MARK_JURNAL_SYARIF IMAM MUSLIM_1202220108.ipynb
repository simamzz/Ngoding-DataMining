### Informasi Dataset

Berikut adalah terjemahan deskripsi dataset ke dalam bahasa Indonesia:

Dataset ini terdiri dari 31 kolom dengan 8337 baris data yang berisikan informasi terkait pelanggan, seperti:

1. **CustomerID**: ID unik yang mengidentifikasi setiap pelanggan.
2. **Count**: Nilai yang digunakan dalam pelaporan/dashboarding untuk menghitung jumlah pelanggan dalam set yang difilter.
3. **Country**: Negara tempat tinggal utama pelanggan.
4. **State**: Negara bagian tempat tinggal utama pelanggan.
5. **City**: Kota tempat tinggal utama pelanggan.
6. **ZipCode**: Kode pos tempat tinggal utama pelanggan.
7. **LatLong**: Gabungan garis lintang dan garis bujur tempat tinggal utama pelanggan.
   a. **Latitude**: Garis lintang tempat tinggal utama pelanggan.
   b. **Longitude**: Garis bujur tempat tinggal utama pelanggan.
8. **Gender**: Jenis kelamin pelanggan: Laki-laki, Perempuan.
9. **Senior Citizen**: Menunjukkan apakah pelanggan berusia 65 tahun atau lebih: Ya, Tidak.
10. **Partner**: Menunjukkan apakah pelanggan memiliki pasangan: Ya, Tidak.
11. **Dependents**: Menunjukkan apakah pelanggan tinggal bersama tanggungan: Ya, Tidak. Tanggungan bisa berupa anak-anak, orang tua, kakek-nenek, dll.
12. **Tenure Months**: Menunjukkan jumlah total bulan pelanggan telah bersama perusahaan hingga akhir kuartal yang ditentukan.
13. **Phone Service**: Menunjukkan apakah pelanggan berlangganan layanan telepon rumah dengan perusahaan: Ya, Tidak.
14. **Multiple Lines**: Menunjukkan apakah pelanggan berlangganan beberapa saluran telepon dengan perusahaan: Ya, Tidak.
15. **Internet Service**: Menunjukkan apakah pelanggan berlangganan layanan Internet dengan perusahaan: Tidak, DSL, Fiber Optic, Kabel.
16. **Online Security**: Menunjukkan apakah pelanggan berlangganan layanan keamanan online tambahan yang disediakan oleh perusahaan: Ya, Tidak.
17. **Online Backup**: Menunjukkan apakah pelanggan berlangganan layanan backup online tambahan yang disediakan oleh perusahaan: Ya, Tidak.
18. **Device Protection**: Menunjukkan apakah pelanggan berlangganan rencana perlindungan perangkat tambahan untuk peralatan Internet mereka yang disediakan oleh perusahaan: Ya, Tidak.
19. **Tech Support**: Menunjukkan apakah pelanggan berlangganan rencana dukungan teknis tambahan dari perusahaan dengan waktu tunggu yang lebih singkat: Ya, Tidak.
20. **Streaming TV**: Menunjukkan apakah pelanggan menggunakan layanan Internet mereka untuk streaming program televisi dari penyedia pihak ketiga: Ya, Tidak. Perusahaan tidak membebankan biaya tambahan untuk layanan ini.
21. **Streaming Movies**: Menunjukkan apakah pelanggan menggunakan layanan Internet mereka untuk streaming film dari penyedia pihak ketiga: Ya, Tidak. Perusahaan tidak membebankan biaya tambahan untuk layanan ini.
22. **Contract**: Menunjukkan jenis kontrak pelanggan saat ini: Bulanan, Satu Tahun, Dua Tahun.
23. **Paperless Billing**: Menunjukkan apakah pelanggan memilih penagihan tanpa kertas: Ya, Tidak.
24. **Payment Method**: Menunjukkan bagaimana pelanggan membayar tagihan mereka: Penarikan Bank, Kartu Kredit, Cek Pos.
25. **Monthly Charge**: Menunjukkan total biaya bulanan pelanggan untuk semua layanan yang mereka peroleh dari perusahaan.
26. **Total Charges**: Menunjukkan total biaya yang ditagihkan kepada pelanggan, dihitung hingga akhir kuartal yang ditentukan.
27. **Churn Label**: Ya = pelanggan meninggalkan perusahaan pada kuartal ini. Tidak = pelanggan tetap bersama perusahaan. Berhubungan langsung dengan Churn Value.
28. **Churn Value**: 1 = pelanggan meninggalkan perusahaan pada kuartal ini. 0 = pelanggan tetap bersama perusahaan. Berhubungan langsung dengan Churn Label.
29. **Churn Score**: Nilai dari 0-100 yang dihitung menggunakan alat prediktif IBM SPSS Modeler. Model ini menggabungkan beberapa faktor yang diketahui menyebabkan churn. Semakin tinggi nilainya, semakin besar kemungkinan pelanggan akan churn.
30. **CLTV**: Customer Lifetime Value. CLTV yang diprediksi dihitung menggunakan formula korporat dan data yang ada. Semakin tinggi nilainya, semakin bernilai pelanggan tersebut. Pelanggan bernilai tinggi harus diawasi untuk potensi churn.
31. **Churn Reason**: Alasan spesifik pelanggan meninggalkan perusahaan. Berhubungan langsung dengan Churn Category.

### 2.3 Missing Value Handling Drop


Menghapus nilai kosong kolom 'Total Charges'

```python
df= df.dropna(subset=['Total Charges'])
```

### 2.4 Missing Value Handling (Mean)


Mengisi nilai kosong kolom 'Tenure Months' dengan nilai mean

```python
df['Tenure Months'] = df['Tenure Months'].fillna(df['Tenure Months'].mean())
```

cek kembali data apakah masih ada nilai kosong setelah value handling mean

```python
cek_mean = df['Tenure Months'].isnull().sum()
print(cek_mean)
```

### 2.5 Missing Value Handling (Median)

Mengisi nilai kosong kolom 'Monthly Charges' dengan nilai median

```python
df['Monthly Charges'] = df['Monthly Charges'].fillna(df['Monthly Charges'].median())
```

cek kembali data apakah masih ada nilai kosong setelah handling value Mean

```python
cek_median = df['Monthly Charges'].isnull().sum()
print(cek_median)
```
