## Studi Kasus Popularitas Penulis Buku Berdasarkan Rating (K-Means Clustering)

### Read Data

Pertama, kita akan membaca dan melihat ringkasan statistik dari dataset menggunakan pandas dan menampilkan beberapa baris awal untuk memahami struktur data.

**Hint: library pandas, menggunkan read_csv(), head(), dan data = df**

```python
df = pd.read_csv("DATA_TP2_K-Means Clustering.csv")

df.head()
```

melihat ringkasan statistik

**Hint: describe()**

```python
df.describe()
```

### Pre-Modelling

Sebelum membangun model, praktikan melakukan pelabelan data untuk memilih fitur yang dijadikan target dan memisahkan data latih dan data uji sesuai ketentuan

#### Labelling Data

Aggregate data berdasarkan nama penulis untuk menghitung jumlah buku (book_count), rata-rata rating (avg_rating), rata-rata skor pada Goodreads (goodreads_avg_rating), dan total rating (total_ratings).

(**Buat dataframe baru hasil aggregasi**)

**Hint: df.groupby().agg(), count, mean, sum, .reset_index()**

```python
aggregated_df = df.groupby('Author').agg(
    book_count=('Book Name', 'count'),
    avg_rating=('Average Rating', 'mean'),
    goodreads_avg_rating=('Score on Goodreads', 'mean'),
    total_ratings=('Number of Ratings', 'sum')
).reset_index()

print(aggregated_df)
```

Rename kolom pada dataframe hasil aggregasi sesuai urutan aggregasi kalian di atas

**Hint: Your_column_name.column =**

```python
aggregated_df.columns = ['Author', 'Book Count', 'Average Rating', 'Goodreads Average Rating', 'Total Ratings']

print(aggregated_df)

# print(df[['Author', 'Book Count', 'Average Rating', 'Goodreads Average Rating', 'Total Ratings']])
```

#### Menentukan K (Elbow Method)

Gunakan Elbow Method untuk menemukan jumlah klaster yang optimal

Import library yang digunakan

**Hint: sklearn.cluster (KMeans),  matplotlib.pyplot (plt)**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

Buat elbow method untuk menentukan k optimal

**Hint: KMeans(), fit(), inertia.append()**

```python
inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(aggregated_df[['Average Rating', 'Total Ratings']])
    # kmeans.fit(x_scale)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

Visualisasikan plot Elbow Method

**Hint: plot(), xlabel(), ylabel(), title()**

```python
import matplotlib.pyplot as plt

plt.plot(range(1, 11), inertia, marker='o', linestyle='--', color='b')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

### Build Model

Praktikan membangun model K-Means dengan data yang sudah dipersiapkan pada tahap sebelumnya

Cocokkan model K-means dengan k optimal (misalnya, k=2)

**Hint: KMeans().fit()**

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(aggregated_df[['Average Rating', 'Total Ratings']])
# kmeans.fit(x)
aggregated_df['Cluster'] = kmeans.labels_

print(aggregated_df)
```

Tetapkan label cluster untuk Author

**Hint: .labels_**

```python
aggregated_df['Cluster'] = kmeans.labels_
print(aggregated_df)
```

Membuat prediksi dari model yang sudah dilatih

**Hint: .predict()**

```python
new_data = pd.DataFrame({'Average Rating': [4.5], 'Total Ratings': [100]})
predictions = kmeans.predict(new_data)

print(predictions)
```

### Evaluate the model (before tuning)

Setelah membangun model dengan melatih dan memprediksi, selanjutnya praktikan mengevaluasi model dengan Visualisasi Hasil Klasterisasi dan Silhouette Score

Import library yang digunakan

**Hint: sklearn.metrics (silhouette_score), sklearn.decomposition (PCA), seaborn (sns)**

```python
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

Untuk visualisasi hasil, praktikan bisa menggunakan PCA (Principal Component Analysis) untuk mereduksi dimensi data sehingga bisa divisualisasikan dalam bentuk 2D.

**Hint: PCA(), pca.fit_transform()**

```python
X = aggregated_df[['Average Rating', 'Total Ratings']]

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis', s=50)

plt.title('Visualisasi Hasil Clustering dengan PCA')
plt.xlabel('PCA Komponen 1')
plt.ylabel('PCA Komponen 2')

plt.colorbar(label='Cluster Label')
plt.show()
```

Buat dataframe hasil PCA untuk visualisasi

**Hint: DataFrame(), columns=[], pca_df['Cluster'] =**



```python
X = aggregated_df[['Average Rating', 'Total Ratings']]

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
pca_df['Cluster'] = kmeans.labels_

print(pca_df.head())
```

Visualisasi plot klusterisasi

**Hint: scatterplot(), hue=, data=pca_df**

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=pca_df, palette='viridis', s=100, marker='o')

plt.title('Visualisasi Hasil Clustering dengan PCA', fontsize=16)
plt.xlabel('PCA1', fontsize=12)
plt.ylabel('PCA2', fontsize=12)

plt.legend(title='Cluster')
plt.show()
```

Menghitung Silhouette Score

**Hint: silhouette_score(), print()**

```python
sil_score = silhouette_score(aggregated_df[['Average Rating', 'Total Ratings']], kmeans.labels_)

print(f'Silhouette Score: {sil_score}')
```

### Model Tuning


Melakukan tuning model dengan memilih hyperparameter secara otomatis menggunakan Grid Search

**Implementasi Grid Search**

Mengetahui nilai mana yang harus ditetapkan untuk parameter yang dicari memerlukan kombinasi pengetahuan dan praktik domain.

Import library yang digunakan

**Hint:sklearn.model_selection (GridSearchCV)**

```python
from sklearn.model_selection import GridSearchCV
```

Tentukan parameter yang ingin dicoba, buat list nilai n_cluster, init, n_init, max_iter pada param_grid

**Hint: param_grid = {n_cluster, ..., ...}**

```python
param_grid = {
    'n_clusters': [2, 3, 4, 5, 6],
    'init': ['k-means++', 'random'],
    'n_init': [10, 15, 20],
    'max_iter': [100, 200, 300]
}
```

Inisialisasi GridSearchCV

**Hint: inisiasi KMeans() dan GridSearchCV (param_grid=..)**

```python
kmeans = KMeans()

grid_search = GridSearchCV(estimator=kmeans, param_grid=param_grid, cv=5, verbose=1)
```

Fit model dengan data training

**Hint: grid_search.fit()**

```python
grid_search.fit(X_train)
# x_sc
```

### Evaluate the model (after tuning)

Tampilkan hasil (parameter dan silhouute score) dari Grid Search


```python
# Jangan ubah apapun pada cell ini
best_score = -1
best_params = None
```

Lengkapi kode dibawah ini

**Hint: Ganti semua 'None' dengan kode yang sesuai**

```python
for params in grid_search.cv_results_['params']:
    model = KMeans(**params)
    cluster_labels = model.fit_predict(aggregated_df)
    score = silhouette_score(aggregated_df, cluster_labels)

    if score > best_score:
        best_score = score
        best_params = params
```

Print hasil dari Grid Search untuk parameter dan skor terbaik

**Hint: best_params, best_score**

```python
print("Best Parameters:", best_params)
print("Best Silhouette Score:", best_score)
```

Berikan kesimpulan berdasarkan hasil evaluasi model (sebelum dan sesudah tuning):

Jawab Disini : 
Berdasarkan hasil evaluasi model K-Means sebelum dan sesudah tuning, dapat disimpulkan bahwa jumlah klaster optimal berada pada 2 atau 3 klaster, sesuai dengan hasil metode Elbow yang menunjukkan titik "siku" pada angka tersebut. Visualisasi PCA setelah tuning juga menunjukkan pemisahan yang cukup jelas antara klaster, yang mengindikasikan bahwa model mampu mengelompokkan data dengan baik. Tuning parameter melalui GridSearchCV membantu meningkatkan akurasi model dengan memilih parameter yang paling optimal, sehingga hasil klaster menjadi lebih konsisten dan akurat.
